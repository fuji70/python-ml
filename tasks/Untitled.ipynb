```python
import numpy as np
import matplotlib.pyplot as plt
import math

def my_func(x):
    return math.sin(x)

# generate data
x = np.arange(100)
y = list(map(lambda x: my_func(x), x))

plt.clf()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x,y)
ax.set_title('first scatter plot')

plt.show()
```
